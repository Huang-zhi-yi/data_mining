## 读出表格

## 清洗数据

### 工作城市栏位处理
* 处理原因：由于工作城市栏位部分数据包含下级区域，为了后边作图便利，需要将下级区域去除掉。
* 数据观察：浏览该栏位发现城市与其下级区域之间使用-分隔，可以以此符号进行处理

#### 筛选出包含下级区域的城市
* 筛选方法:利用contains把包含-符号的数据筛选出来

```python

包含下级区域的城市=df[df["工作城市"].str.contains("-")]
包含下级区域的城市

```

#### 生成列表准备处理
* 取筛选后的结果的工作城市一列并生成列表

```python

包含下级区域的城市["工作城市"].to_list()
```

#### 利用for循环处理数据
* .split("-")[0]以“-”字符为切点，切分成列表，然后[0]取出第一个元素
* .append()添加到新的列表中

```python

处理后的列表=list()
for x in 包含下级区域的城市["工作城市"].to_list():
        处理后的列表.append(x.split("-")[0])  
处理后的列表
```

#### 数据原表替换
* 将处理完的数据列表替换掉原来的工作城市一列

```python
# 把切除干净后的列表替代原先一列
包含下级区域的城市["工作城市"]=处理后的列表
包含下级区域的城市
```

#### 数据更新
* 将处理完的数据重新合并到原有表格

```python

df=df.drop(包含下级区域的城市.index)
df
df=df.append(包含下级区域的城市,ignore_index=True) 
df
```

#### 进行数据的分进合击，为做图做准备

```python

教育岗位城市招聘数量=df.groupby("工作城市").agg({"职位":"count"})
教育岗位城市招聘数量
```

```python
# 重新设置索引，生成数据列表，再做成区域地图。
教育岗位城市招聘数量=教育岗位城市招聘数量.reset_index()
工作城市数量列表=教育岗位城市招聘数量["工作城市"].to_list()
工作职位数量列表=教育岗位城市招聘数量["职位"].to_list()
```

## 职位要求主题建模
* 为什么要主题建模:招聘要求往往有共性与个性，建立主题模型，可以让求职者了解到这些公司招聘教育岗位最看中什么，然后可以帮助求职者根据自身实际有针对投放简历。

### 清洗数据
* 原因：由于数据中会出现任职要求、岗位要求这些比较杂的数据，因此需要对数据进行清洗，提交主题建模的准确度

#### 利用split、join()对无关数据清洗

```python
# 无关数据
职位描述列表=df["职位描述"].to_list()
职位描述列表
合并职位描述="".join([a.strip() for a in 职位描述列表])
合并职位描述
ccs="".join(合并职位描述.split("任职要求："))
数据处理后=" ".join(ccs.split("岗位职责："))#.split("工作职责：")
二次处理="".join(数据处理后.split("\t"))

information=二次处理.split("\uf06c") 
information
```

#### 读入过滤词表格

```python
import jieba
fn = { "input" : { "submissions": "data_sets/submissions_profit_model_data_.xlsx",\
                   "user_dict_jieba": "data_sets/user_dict_jieba_expanded.txt",\
                   "zh_stopwords": "data_sets/stopwords_zh_both.txt",\
                   "word_freq_manual": "data_sets/submissions_profit_model_word_freq_手动.xlsx"
                 } 
     }
```

```python
stop = [line.strip() for line in open(fn["input"]["zh_stopwords"],"r",encoding="UTF-8-sig").readlines()]
stop
```

### 建立函数并利用jieba分词模块过滤停止词并进行分词 

```python
fout = open('断词后结果.txt', 'w')
def splitSentence(inputFile):
    for eachLine in inputFile:
        line = eachLine.strip()#.decode('utf-8', 'ignore')      #去除每行首尾可能出现的空格，并转为Unicode进行处理
        line=line.strip('\n')                                   #去掉多余空行
        wordList = list(jieba.cut(line))
        outStr = ''
        for word in wordList:#
            if len(word)>1:
                if not word in stop:
                        outStr += word
                        outStr += ' '
        fout.write(str(outStr.strip()))              #将分词好的结果写入到输出文件.encode('utf-8')
        fout.write('\n')
```

```python
splitSentence(information)
```

### 读入分词后文档准备做数据主题建模

```python
with open('断词后结果.txt' ,'r') as write:
    text=write.read()
    print(text)
```

#### 根据后边建模的要求生成相应的列表

```python
text1=[text.split(" ")]# 嵌套列表
text1
word=text.split(" ")#列表
word
```

### 主题建模
* 使用模块:[gensim](https://radimrehurek.com/gensim/)

```python
from gensim.corpora import Dictionary
import gensim

from gensim import corpora, models
from gensim.corpora.dictionary import Dictionary
from gensim.models.ldamodel import LdaModel
dictionary = Dictionary(text1)#嵌套词语列表
print((dictionary))


doc = [dictionary.doc2bow(word)]#词语列表
print(doc)
```

```python
#num_topics是建模的主题数； num_words是建模的主题词显示个数
ldamodel =LdaModel(corpus=doc, num_topics=5,id2word=dictionary)
lda=ldamodel.print_topics(num_topics=5, num_words=30)
lda
```
